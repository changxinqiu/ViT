```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches

# --- 1.  ( fenics_solver.py ) ---
L = 5.0        # 
H = 1.0        # 
x1 = 1.0       # 
x2 = 4.0       # 
h_D = 0.4      # 

# --- 2.  ---
fig, ax = plt.subplots(figsize=(10, 4), dpi=150)

# --- 3.  ---

# A.  ( Stokes )
# 
domain_rect = patches.Rectangle((0, 0), L, H, linewidth=1, edgecolor='black', facecolor='#E0F7FA', label='Stokes Region ($\Omega_S$)')
ax.add_patch(domain_rect)

# B.  Darcy  ()
# x  x1  x2y  0  h_D
darcy_rect = patches.Rectangle((x1, 0), x2-x1, h_D, linewidth=1, edgecolor='black', facecolor='#FFECB3', hatch='...', label='Darcy Region ($\Omega_D$)')
ax.add_patch(darcy_rect)

# --- 4.  (Interface) ---
#  Darcy  Stokes 
#  ()
#  FEniCS 
# 

# 
plt.plot([x1, x2], [h_D, h_D], color='red', linewidth=2.5, linestyle='--', label='Interface ($\Gamma$)')
# 
plt.plot([x1, x1], [0, h_D], color='red', linewidth=2.5, linestyle='--')
# 
plt.plot([x2, x2], [0, h_D], color='red', linewidth=2.5, linestyle='--')

# --- 5.  (BCs) ---

#  (Inlet) -  ()
plt.arrow(-0.3, (H+h_D)/2, 0.2, 0, width=0.05, head_width=0.15, color='blue')
plt.text(-0.4, (H+h_D)/2, 'Inlet\n($u_{in}$)', ha='right', va='center', color='blue', fontsize=10)

#  (Outlet) - 
plt.arrow(L+0.1, (H+h_D)/2, 0.2, 0, width=0.05, head_width=0.15, color='green')
plt.text(L+0.4, (H+h_D)/2, 'Outlet\n($p=0$)', ha='left', va='center', color='green', fontsize=10)

#  (Walls)
plt.text(L/2, H+0.05, 'No-slip Wall', ha='center', va='bottom', fontsize=9)
plt.text(L/2, -0.1, 'Bottom Wall', ha='center', va='top', fontsize=9)

# --- 6.  ---
# 
ax.text(L/2, H*0.7, r'$\Omega_S$ (Stokes Flow)', ha='center', va='center', fontsize=14, color='#006064', fontweight='bold')
ax.text((x1+x2)/2, h_D/2, r'$\Omega_D$ (Porous Bed)', ha='center', va='center', fontsize=12, color='#BF360C', fontweight='bold')

# 
plt.scatter([x1, x2], [0, 0], color='black', s=20, zorder=10)
ax.text(x1, -0.05, f'$x_1={x1}$', ha='center', va='top', fontsize=8)
ax.text(x2, -0.05, f'$x_2={x2}$', ha='center', va='top', fontsize=8)
ax.text(-0.05, h_D, f'$h_D={h_D}$', ha='right', va='center', fontsize=8)

# --- 7.  ---
ax.set_xlim(-0.5, L+0.5)
ax.set_ylim(-0.2, H+0.2)
ax.set_aspect('equal')
ax.set_title('Computational Domain: Catalyst Bed Reactor (Stokes-Darcy Coupled)', fontsize=12, pad=15)
ax.legend(loc='upper right', bbox_to_anchor=(1.0, 1.25), ncol=3, frameon=False)
plt.xlabel('x (m)')
plt.ylabel('y (m)')

# 
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['left'].set_visible(False)
ax.spines['bottom'].set_visible(False)
ax.tick_params(axis='both', which='both', length=0) # 
# 
ax.set_xticks([0, L])
ax.set_yticks([0, H])

plt.tight_layout()
plt.show()

# 
# plt.savefig('catalyst_bed_domain.png', dpi=300, bbox_inches='tight')
```
